```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
# Normalizasyon: (x - mean) / std
# Fashion-MNIST için yaygın değerler:
mean, std = 0.2860, 0.3530

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((mean,), (std,))
])

trainval_ds = datasets.FashionMNIST(root="./data", train=True, download=True, transform=transform)
test_ds     = datasets.FashionMNIST(root="./data", train=False, download=True, transform=transform)

n_total = len(trainval_ds)           # 60000
n_train = int(0.70 * n_total)        # 42000
n_val   = n_total - n_train          # 18000  (bu 70/30 olur)

# İstersek tam 70/15/15 yapmak için test'i de katmak gerekir.
# Ama ödevler genelde resmi 10k test'i korumayı sever.
# İstersen tam 70/15/15'i bir sonraki adımda "train+test'i birleştirip" yapabiliriz.

train_ds, val_ds = random_split(trainval_ds, [n_train, n_val], generator=torch.Generator().manual_seed(42))

len(train_ds), len(val_ds), len(test_ds)
```

```python
batch_size = 128

train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
```

```python
class_names = trainval_ds.classes
class_names
```

```python
# 1 batch alıp 12 örnek gösterelim
images, labels = next(iter(train_loader))

plt.figure(figsize=(10,4))
for i in range(12):
    plt.subplot(2,6,i+1)
    # Normalize edildiği için görüntü "garip" durabilir; burada sadece şekil için gösteriyoruz
    plt.imshow(images[i].squeeze().numpy(), cmap="gray")
    plt.title(class_names[labels[i]])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
class MLP(nn.Module):
    def __init__(self, input_dim=28*28, num_classes=10):
        super().__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.net(x)

model = MLP().to(device)
model
```

```python
def count_params(m):
    return sum(p.numel() for p in m.parameters() if p.requires_grad)

count_params(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
def accuracy_from_logits(logits, y):
    preds = logits.argmax(dim=1)
    return (preds == y).float().mean().item()

@torch.no_grad()
def evaluate(model, loader):
    model.eval()
    total_loss, total_acc, n = 0.0, 0.0, 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        logits = model(x)
        loss = criterion(logits, y)
        bs = x.size(0)
        total_loss += loss.item() * bs
        total_acc  += (logits.argmax(1) == y).float().sum().item()
        n += bs
    return total_loss / n, total_acc / n

def train_one_epoch(model, loader):
    model.train()
    total_loss, total_acc, n = 0.0, 0.0, 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)

        optimizer.zero_grad()
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()

        bs = x.size(0)
        total_loss += loss.item() * bs
        total_acc  += (logits.argmax(1) == y).float().sum().item()
        n += bs

    return total_loss / n, total_acc / n
```

```python
epochs = 10
history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

for epoch in range(1, epochs+1):
    tr_loss, tr_acc = train_one_epoch(model, train_loader)
    va_loss, va_acc = evaluate(model, val_loader)

    history["train_loss"].append(tr_loss)
    history["train_acc"].append(tr_acc)
    history["val_loss"].append(va_loss)
    history["val_acc"].append(va_acc)

    print(f"Epoch {epoch:02d} | "
          f"train loss {tr_loss:.4f} acc {tr_acc:.4f} | "
          f"val loss {va_loss:.4f} acc {va_acc:.4f}")
```

```python
plt.figure()
plt.plot(history["train_loss"], label="train loss")
plt.plot(history["val_loss"], label="val loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(history["train_acc"], label="train acc")
plt.plot(history["val_acc"], label="val acc")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# model 1 başlangıç

@torch.no_grad()
def predict_all(model, loader):
    model.eval()
    all_preds, all_labels = [], []
    for x, y in loader:
        x = x.to(device)
        logits = model(x)
        preds = logits.argmax(dim=1).cpu().numpy()
        all_preds.append(preds)
        all_labels.append(y.numpy())
    return np.concatenate(all_labels), np.concatenate(all_preds)

y_true, y_pred = predict_all(model, test_loader)

print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
fig, ax = plt.subplots(figsize=(8,8))
disp.plot(ax=ax, xticks_rotation=45, values_format="d")
plt.title("Model 1 (MLP) - Confusion Matrix (Test)")
plt.show()
```

```python
import torch
import torch.nn as nn

```

```python
class CNN_Baseline(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()

        # --------- Feature extractor (özellik çıkarıcı) ---------
        # Burada CNN'in asıl farkını yaratan kısım var:
        # Konvolüsyon ile "yerel/uzamsal" özellikleri öğreniyoruz (kenar, köşe, şekil parçaları).
        self.features = nn.Sequential(
            # Conv2d: "küçük pencere" (kernel) ile resmi tarar.
            # in_channels=1  -> Fashion-MNIST gri ton (tek kanal)
            # out_channels=32 -> 32 farklı filtre/özellik haritası öğren
            # kernel_size=3 -> 3x3 pencerelerle tarama
            # padding=1 -> çıktı boyutu 28x28 kalsın (kenarlardan bilgi kaybı azalsın)
            nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1),

            # ReLU: doğrusal olmayanlık ekler. "Öğrenilebilir ifade gücü" sağlar.
            nn.ReLU(),

            # MaxPool2d(2): 2x2 bölgede en büyüğü seçer.
            # 28x28 -> 14x14 düşer (hesap azalır + küçük kaymalara dayanıklılık artar)
            nn.MaxPool2d(kernel_size=2),

            # İkinci konvolüsyon: daha "yüksek seviye" özellikler (kenar birleşimleri, daha kompleks parçalar)
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1),
            nn.ReLU(),

            # 14x14 -> 7x7
            nn.MaxPool2d(kernel_size=2),
        )

        # --------- Classifier (sınıflandırıcı) ---------
        # Özellik haritalarını sınıf skorlarına çevirir.
        self.classifier = nn.Sequential(
            # (batch, 64, 7, 7) -> (batch, 64*7*7)
            nn.Flatten(),

            # Dense katman: özellikleri birleştirip karar vermeyi kolaylaştırır
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),

            # Son katman: 10 sınıf için 10 skor (logits)
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        # x: (batch, 1, 28, 28)
        x = self.features(x)     # -> (batch, 64, 7, 7)
        x = self.classifier(x)   # -> (batch, 10)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model2 = CNN_Baseline().to(device)
model2
```

```python
def count_params(m):
    return sum(p.numel() for p in m.parameters() if p.requires_grad)

print("Trainable params:", count_params(model2))
```

```python
criterion2 = nn.CrossEntropyLoss()
optimizer2 = torch.optim.Adam(model2.parameters(), lr=1e-3)
```

```python
@torch.no_grad()
def evaluate_generic(model, loader, criterion):
    model.eval()  # dropout/bn olsaydı eval davranışı için kritik
    total_loss, correct, n = 0.0, 0, 0

    for x, y in loader:
        x, y = x.to(device), y.to(device)
        logits = model(x)
        loss = criterion(logits, y)

        bs = x.size(0)
        total_loss += loss.item() * bs
        correct += (logits.argmax(1) == y).sum().item()
        n += bs

    return total_loss / n, correct / n


def train_one_epoch_generic(model, loader, criterion, optimizer):
    model.train()
    total_loss, correct, n = 0.0, 0, 0

    for x, y in loader:
        x, y = x.to(device), y.to(device)

        optimizer.zero_grad()      # önceki batch gradyanlarını temizle
        logits = model(x)          # ileri yayılım (forward)
        loss = criterion(logits, y)# kayıp
        loss.backward()            # geri yayılım (backprop): gradyan hesapla
        optimizer.step()           # ağırlıkları güncelle

        bs = x.size(0)
        total_loss += loss.item() * bs
        correct += (logits.argmax(1) == y).sum().item()
        n += bs

    return total_loss / n, correct / n
```

```python
epochs = 10
hist2 = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

for epoch in range(1, epochs + 1):
    tr_loss, tr_acc = train_one_epoch_generic(model2, train_loader, criterion2, optimizer2)
    va_loss, va_acc = evaluate_generic(model2, val_loader, criterion2)

    hist2["train_loss"].append(tr_loss)
    hist2["train_acc"].append(tr_acc)
    hist2["val_loss"].append(va_loss)
    hist2["val_acc"].append(va_acc)

    print(f"Epoch {epoch:02d} | "
          f"train loss {tr_loss:.4f} acc {tr_acc:.4f} | "
          f"val loss {va_loss:.4f} acc {va_acc:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(hist2["train_loss"], label="train loss")
plt.plot(hist2["val_loss"], label="val loss")
plt.xlabel("epoch"); plt.ylabel("loss")
plt.title("Model 2 (CNN) Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(hist2["train_acc"], label="train acc")
plt.plot(hist2["val_acc"], label="val acc")
plt.xlabel("epoch"); plt.ylabel("accuracy")
plt.title("Model 2 (CNN) Accuracy")
plt.legend()
plt.show()
```
